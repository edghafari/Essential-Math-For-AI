### **1. Linear Algebra (The Language of Data)**

Linear algebra is arguably the most critical mathematical field for AI/ML. It provides the framework for representing and manipulating data in a way that computers can process efficiently. At its core, AI is about processing data, and linear algebra is the language used to express that data.

#### **Why It's Essential**

Data in AI is almost always represented as a numerical structure. A grayscale image is a matrix of pixel intensities, a user's profile is a vector of features, and a collection of documents can be represented as a document-term matrix. Linear algebra gives us the tools to perform operations on these structures at scale. Neural networks, the engine of modern deep learning, are fundamentally a series of interconnected linear algebra operations.

#### **Key Topics Explained**

* **Vectors and Matrices:**
    * **What they are:** Vectors are arrays of numbers that can represent a point in space or a set of features (e.g., a user's age, income, and location). Matrices are grids of numbers that can represent an entire dataset (rows for samples, columns for features), a grayscale image, or a transformation.
    * **Operations:** Matrix multiplication is the cornerstone of neural networks. Each layer's computation is essentially an input matrix multiplied by a weight matrix. Inverses and determinants are crucial for solving systems of linear equations, which appear in models like Linear Regression.

* **Eigenvalues and Eigenvectors:**
    * **Intuition:** Eigenvectors of a matrix are special vectors that do not change their direction when a linear transformation is applied to them; they are only scaled by a factor called the eigenvalue.
    * **Application (PCA):** In Principal Component Analysis (PCA), a popular dimensionality reduction technique, eigenvectors of the data's covariance matrix represent the principal components (the directions of maximum variance). The eigenvalues indicate the amount of variance captured by each component. By keeping only the eigenvectors with the highest eigenvalues, you can reduce the data's dimensionality while retaining most of its important information.

* **Matrix Factorization:**
    * **Concept:** This is the process of breaking down a large matrix into smaller, constituent matrices.
    * **Application (Recommendation Systems):** In a system like Netflix's, you have a massive, sparse matrix of users and their movie ratings. Matrix factorization techniques like Singular Value Decomposition (SVD) can decompose this into a "user-feature" matrix and a "movie-feature" matrix. These lower-dimensional "feature" vectors (or embeddings) capture latent tastes (e.g., affinity for sci-fi, preference for a specific director). To predict a rating, you simply take the dot product of a user's vector and a movie's vector.

* **Tensors:**
    * **Concept:** Tensors are a generalization of matrices to higher dimensions. A 0D tensor is a scalar, a 1D tensor is a vector, and a 2D tensor is a matrix. A color image is a 3D tensor (height x width x 3 color channels), and a batch of such images would be a 4D tensor.
    * **Application:** Deep learning frameworks like TensorFlow and PyTorch are built around tensors. All operations, from simple additions to complex convolutions, are performed on tensors.

### **2. Calculus (The Engine of Optimization)**

If linear algebra provides the structure for data, calculus provides the tools to learn from it. The process of "training" a machine learning model is fundamentally an optimization problem: finding the model parameters that result in the best performance. Calculus is the key to this optimization process.

#### **Why It's Very Important**

Training involves adjusting a model's parameters (like the weights in a neural network) to minimize a "loss function," which measures how wrong the model's predictions are. Calculus, specifically differentiation, tells us how to make these adjustments. It allows us to see how a tiny change in a parameter will affect the overall error, guiding us toward the best set of parameters.

#### **Key Topics Explained**

* **Differentiation (Partial Derivatives and Gradients):**
    * **Concept:** The derivative of a function measures its rate of change. In ML, we use partial derivatives to see how the loss function changes with respect to *each individual parameter* in the model. The **gradient** is simply a vector containing all of these partial derivatives.
    * **Application (Backpropagation):** The gradient tells you the direction of the steepest ascent of the loss function. To minimize the loss, we move in the *opposite* direction of the gradient. This is the core idea behind **Gradient Descent**. Backpropagation is the clever algorithm that efficiently calculates the gradient for every parameter in a deep neural network, layer by layer, starting from the output and moving backward.

* **Multivariable Calculus:**
    * **Relevance:** Since models have thousands or even millions of parameters, the loss function is a function of many variables. Understanding multivariable calculus is essential to navigate this high-dimensional "loss landscape" and find its minimum point.

* **Optimization Techniques:**
    * **Gradient Descent:** Imagine you are on a mountain in a thick fog and want to get to the lowest valley. You can feel the slope of the ground beneath your feet (the gradient). The most straightforward strategy is to take a step in the steepest downward direction. You repeat this process until you can no longer go down. This is Gradient Descent.
    * **Stochastic Gradient Descent (SGD):** In standard Gradient Descent, you calculate the gradient using the entire dataset, which is computationally expensive. In SGD, you estimate the gradient using just a small, random subset (a "mini-batch") of the data. This is much faster and is the standard method for training deep neural networks.

### **3. Probability and Statistics (The Foundation of Uncertainty and Inference)**

AI and machine learning are fundamentally about making predictions and decisions under uncertainty. Probability theory provides the framework for quantifying this uncertainty, while statistics offers the tools to analyze data, draw inferences, and validate models.

#### **Why It's Essential**

Data is never perfect; it's noisy and incomplete. Probability allows us to build models that don't just give a single, deterministic answer but rather a distribution of possible outcomes. Statistics are essential for everything from understanding your data's characteristics to rigorously evaluating whether your model's performance is statistically significant or just due to chance.

#### **Key Topics Explained**

* **Probability Distributions:**
    * **Concept:** These are mathematical functions that describe the likelihood of different outcomes.
    * **Examples:** The **Normal (Gaussian) distribution** is used to model many natural phenomena and is a common assumption for data noise. The **Binomial distribution** can model the probability of a certain number of "successes" in a set number of trials (e.g., predicting how many users will click an ad).

* **Bayes' Theorem:**
    * **Formula:** $P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}$
    * **Intuition:** It describes how to update your belief in a hypothesis ($A$) given new evidence ($B$). It's a formal way of learning from data.
    * **Application (Naive Bayes):** The Naive Bayes classifier uses this theorem to calculate the probability that a given data point belongs to a certain class. For example, in spam detection, it calculates the probability of an email being spam given the words it contains.

* **Expectation, Variance, and Covariance:**
    * **Concepts:** **Expectation** is the average value of a random variable. **Variance** measures how spread out the data is from the mean. **Covariance** and **Correlation** measure how two variables move together.
    * **Application:** These concepts are used everywhere, from initializing weights in a neural network to understanding the relationships between features in your dataset.

* **Maximum Likelihood Estimation (MLE):**
    * **Concept:** MLE is a method for estimating the parameters of a statistical model. It asks: "What parameters for this model would make the observed data most probable?"
    * **Application:** It's the underlying principle for how many ML models are fit to data, including logistic regression.

1. Linear Algebra: Principal Component Analysis (PCA)
This example shows how eigenvectors are used to find the principal components of a dataset. These components represent the directions of maximum variance and are fundamental to dimensionality reduction.

```python
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Create correlated 2D data ---
# We create data that has a clear linear trend
np.random.seed(42)
X = np.dot(np.random.rand(2, 2), np.random.randn(2, 200)).T

# --- 2. Center the data ---
# Subtract the mean to center the data around the origin
mean = np.mean(X, axis=0)
X_centered = X - mean

# --- 3. Calculate covariance matrix and eigenvectors ---
# The covariance matrix describes the variance and correlation of the features
cov_matrix = np.cov(X_centered, rowvar=False)
# Eigenvectors give the directions of variance, eigenvalues give the magnitude
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# --- 4. Plotting ---
plt.figure(figsize=(8, 8))
# Plot the centered data points
plt.scatter(X_centered[:, 0], X_centered[:, 1], alpha=0.7, label='Centered Data Points')

# Plot the eigenvectors (the principal components)
# We scale them by their eigenvalue for better visualization
for i in range(len(eigenvalues)):
    eigenvector = eigenvectors[:, i]
    # The arrow starts at the origin (0,0) and points in the eigenvector's direction
    plt.arrow(0, 0, eigenvector[0] * np.sqrt(eigenvalues[i]),
              eigenvector[1] * np.sqrt(eigenvalues[i]),
              head_width=0.2, head_length=0.2, fc='r', ec='r',
              label=f'Principal Component {i+1}' if i==0 else '_nolegend_')

plt.title('Linear Algebra: Principal Component Analysis', fontsize=16)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()
plt.axis('equal') # Ensure aspect ratio is equal for accurate angle representation
plt.show()

print("Eigenvectors (Principal Components):\n", eigenvectors)
print("\nEigenvalues (Magnitude of Variance):\n", eigenvalues)
```

Explanation: The red arrows are the eigenvectors of the data's covariance matrix. Notice how the first principal component (the longer arrow) points along the direction of the greatest spread in the data. PCA uses these vectors to transform the data, reducing its dimensions while preserving the most important information.

2. Calculus: Gradient Descent
This example visually demonstrates Gradient Descent, the core optimization algorithm for training models. We find the minimum of a simple function by iteratively taking steps in the opposite direction of the gradient.

```python
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Define the function and its derivative (gradient) ---
# This is our "loss function" we want to minimize
def loss_function(x):
    return x**2 - 4*x + 5

# The derivative tells us the slope at any point
def gradient(x):
    return 2*x - 4

# --- 2. Implement Gradient Descent ---
learning_rate = 0.1
n_iterations = 30
current_x = -1.5  # A random starting point
path = [current_x]

for _ in range(n_iterations):
    grad = gradient(current_x)
    current_x = current_x - learning_rate * grad # Move in the opposite direction of the gradient
    path.append(current_x)

path = np.array(path)

# --- 3. Plotting ---
x_vals = np.linspace(-2, 6, 100)
y_vals = loss_function(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='Loss Function: $f(x) = x^2 - 4x + 5$')
plt.plot(path, loss_function(path), 'ro-', label='Path of Gradient Descent')
plt.title('Calculus: Gradient Descent Optimization', fontsize=16)
plt.xlabel('Parameter value (x)')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()

print(f"The minimum found by Gradient Descent is at x = {path[-1]:.4f}")
```

Explanation: The plot shows the algorithm starting at x = -1.5 and taking steps "downhill" along the curve. Each step is guided by the derivative (the gradient). The steps get smaller as the slope flattens, eventually converging to the function's minimum. This is exactly how a neural network learns to minimize its prediction error.

3. Probability and Statistics: The Normal Distribution
Machine learning models often assume that data or the errors in predictions follow a certain statistical distribution. This example generates data from a Normal (Gaussian) distribution and shows how a histogram of the data approximates the theoretical distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# --- 1. Generate data from a Normal distribution ---
mu, sigma = 0, 10 # Mean and standard deviation
np.random.seed(0)
s = np.random.normal(mu, sigma, 1000) # Generate 1000 samples

# --- 2. Plotting ---
plt.figure(figsize=(10, 6))

# Plot a histogram of the generated data
sns.histplot(s, bins=30, kde=False, stat='density', label='Generated Data Histogram')

# Overlay the theoretical Probability Density Function (PDF)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, sigma)
plt.plot(x, p, 'k', linewidth=2, label='Theoretical PDF')

plt.title(f'Probability: Normal Distribution ($\mu={mu}, \sigma={sigma}$)', fontsize=16)
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

Explanation: The blue bars represent the frequency of the randomly generated data points. The black line is the perfect, theoretical Normal distribution curve. As you can see, the shape of the data closely follows the theoretical distribution, illustrating how we can model real-world data using statistical functions.

4. Optimization: Comparing Optimizers
This example visualizes the paths taken by two different gradient-based optimizersâ€”standard Gradient Descent (GD) and GD with Momentumâ€”on a 2D loss surface. It shows why more advanced optimizers can be more efficient.

```python
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Define a 2D loss surface (a skewed bowl shape) ---
def loss_surface(x, y):
    return x**2 + 10 * y**2

def gradient_surface(x, y):
    return np.array([2*x, 20*y])

# --- 2. Implement the optimizers ---
def run_optimizer(optimizer_type='gd', n_iterations=20):
    x, y = 4.5, 2.0  # Starting point
    path = [(x, y)]
    velocity = np.zeros(2) # For momentum
    lr = 0.08 if optimizer_type == 'gd' else 0.04 # Learning rates tuned for stability
    beta = 0.8 # Momentum factor

    for _ in range(n_iterations):
        grad = gradient_surface(x, y)
        if optimizer_type == 'momentum':
            velocity = beta * velocity + (1 - beta) * grad
            update = velocity
        else: # Standard GD
            update = grad

        x, y = x - lr * update[0], y - lr * update[1]
        path.append((x, y))
    return np.array(path)

path_gd = run_optimizer('gd')
path_momentum = run_optimizer('momentum')

# --- 3. Plotting ---
x = np.linspace(-5, 5, 100)
y = np.linspace(-2.5, 2.5, 100)
X, Y = np.meshgrid(x, y)
Z = loss_surface(X, Y)

plt.figure(figsize=(10, 7))
plt.contour(X, Y, Z, levels=np.logspace(0, 3, 10), cmap='viridis')
plt.plot(path_gd[:, 0], path_gd[:, 1], 'o-', label='Standard Gradient Descent', color='red')
plt.plot(path_momentum[:, 0], path_momentum[:, 1], 'o-', label='GD with Momentum', color='blue')
plt.title('Optimization: Comparing Optimizer Paths', fontsize=16)
plt.xlabel('Parameter 1 (x)')
plt.ylabel('Parameter 2 (y)')
plt.legend()
plt.grid(True)
plt.show()
```

Explanation: The background contours represent the "loss landscape" we want to navigate. Notice that Standard GD (red) takes a noisy, zig-zagging path because the gradients in the y-direction are much steeper. GD with Momentum (blue) smooths out these oscillations by accumulating past gradients, resulting in a much more direct and efficient path to the minimum. This is why optimizers like Adam and RMSprop are preferred in practice.

5. Discrete Mathematics: Graph Theory
This example uses the networkx library to model and visualize a simple graph, like a social network. Graph theory is essential for Graph Neural Networks (GNNs), recommendation engines, and knowledge graphs.

```python
import networkx as nx
import matplotlib.pyplot as plt

# --- 1. Create a graph ---
G = nx.Graph()

# Add nodes (e.g., people)
people = ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace']
G.add_nodes_from(people)

# Add edges (e.g., friendships)
edges = [('Alice', 'Bob'), ('Alice', 'Charlie'), ('Bob', 'Charlie'),
         ('Bob', 'David'), ('Charlie', 'Eve'), ('David', 'Eve'),
         ('Eve', 'Frank'), ('Frank', 'Grace'), ('Eve', 'Grace')]
G.add_edges_from(edges)

# --- 2. Calculate a centrality measure ---
# Degree centrality is the number of edges connected to a node
centrality = nx.degree_centrality(G)

# --- 3. Plotting ---
plt.figure(figsize=(10, 8))
node_sizes = [v * 2000 for v in centrality.values()] # Size nodes by centrality
node_colors = [centrality[node] for node in G.nodes()] # Color nodes by centrality

pos = nx.spring_layout(G, seed=42)  # Positions for all nodes
nx.draw(G, pos, with_labels=True, node_size=node_sizes,
        node_color=node_colors, cmap=plt.cm.plasma, font_size=12, font_color='white')

plt.title('Discrete Math: Social Network Graph', fontsize=16)
plt.show()

print("Node with highest centrality (most connections):", max(centrality, key=centrality.get))
```

Explanation: The plot shows a network of individuals where lines represent friendships. The nodes are sized and colored by their "degree centrality"â€”how many connections they have. You can visually identify 'Eve' as a key connector in this network. GNNs leverage this structural information to make predictions, such as recommending new friends or classifying nodes.

6. Information Theory: KL Divergence
Kullback-Leibler (KL) Divergence measures how one probability distribution differs from a second, reference distribution. It's a measure of "surprise" or "information loss." It is fundamental to many loss functions in deep learning.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, entropy

# --- 1. Define two different Normal distributions ---
x = np.linspace(-10, 10, 1000)
# Distribution P (our reference)
p_dist = norm.pdf(x, loc=0, scale=2)
# Distribution Q (the one we are comparing to P)
q_dist = norm.pdf(x, loc=2, scale=2.5)

# --- 2. Calculate KL Divergence ---
# We need to ensure the values are normalized to be a probability distribution
p_dist_norm = p_dist / np.sum(p_dist)
q_dist_norm = q_dist / np.sum(q_dist)

# entropy(pk, qk) calculates KL(P || Q)
kl_pq = entropy(p_dist_norm, qk=q_dist_norm)
kl_qp = entropy(q_dist_norm, qk=p_dist_norm)


# --- 3. Plotting ---
plt.figure(figsize=(10, 6))
plt.plot(x, p_dist_norm, 'b-', label='Distribution P (Reference)')
plt.plot(x, q_dist_norm, 'r--', label='Distribution Q (Approximation)')
plt.fill_between(x, p_dist_norm, alpha=0.2, color='blue')
plt.fill_between(x, q_dist_norm, alpha=0.2, color='red')

plt.title('Information Theory: KL Divergence', fontsize=16)
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
plt.text(-9, 0.003, f'KL(P || Q) = {kl_pq:.4f}\nKL(Q || P) = {kl_qp:.4f}',
         bbox=dict(facecolor='white', alpha=0.8))
plt.grid(True)
plt.show()
```

Explanation: The plot shows two different probability distributions. KL Divergence quantifies the "gap" between them. In AI, P might be the true data distribution, and Q might be the distribution generated by our model. The goal of training is often to minimize this divergence, making our model's output as close to reality as possible. Note that KL(P || Q) is not the same as KL(Q || P), meaning it's an asymmetric measure. The cross-entropy loss function used in classification is directly related to KL Divergence.

1. Vectors and Matrices: The Building Blocks
This example covers basic vector and matrix operations using NumPy. These operations are the foundation of almost every AI model, from simple linear regression to complex neural networks.

Concept: Vectors represent points or features, while matrices represent datasets or linear transformations. Matrix multiplication is how a neural network layer processes input data.

```python
import numpy as np

# --- Vector Operations ---
print("--- Vector Operations ---")
# Create two vectors (1D arrays)
v = np.array([1, 2, 3])
u = np.array([4, 5, 6])

print(f"Vector v: {v}")
print(f"Vector u: {u}")

# Vector Addition
print(f"v + u = {v + u}")

# Dot Product (a fundamental operation in calculating weighted sums)
print(f"Dot product of v and u: {np.dot(v, u)}")


# --- Matrix Operations ---
print("\n--- Matrix Operations ---")
# Create two matrices (2D arrays)
# Matrix A could represent a small dataset (2 samples, 3 features)
A = np.array([[1, 2, 3],
              [4, 5, 6]])

# Matrix B could represent the weights of a neural network layer
B = np.array([[7, 8],
              [9, 10],
              [11, 12]])

print(f"Matrix A (2x3):\n{A}")
print(f"Matrix B (3x2):\n{B}")

# Matrix Multiplication (dot product)
# This is the core operation of a neural network layer
C = np.dot(A, B)
print(f"Matrix Multiplication (A @ B):\n{C}")

# --- Inverse and Determinant ---
print("\n--- Inverse and Determinant ---")
# These are only defined for square matrices
M = np.array([[2, 4],
              [1, 3]])

# Determinant: A scalar value that gives properties of the matrix's transformation
# A determinant of 0 means the matrix is "singular" and has no inverse.
det_M = np.linalg.det(M)
print(f"Matrix M:\n{M}")
print(f"Determinant of M: {det_M:.2f}")

# Inverse: A matrix that, when multiplied by M, results in the identity matrix.
# Used for solving systems of linear equations.
inv_M = np.linalg.inv(M)
print(f"Inverse of M:\n{inv_M}")

# Verification: M * M_inv should be the identity matrix [[1, 0], [0, 1]]
identity = np.dot(M, inv_M)
print(f"Verification (M @ M_inv):\n{np.round(identity)}")
```

2. Eigenvalues and Eigenvectors: Directions of Transformation
This example demonstrates how to find eigenvalues and eigenvectors and visualizes their meaning. Eigenvectors are the vectors that a linear transformation only scales, without changing their direction. This is the core concept behind PCA for dimensionality reduction.

Concept: In PCA, the eigenvectors of the covariance matrix are the "principal components" (directions of maximum variance), and the eigenvalues indicate how much variance is captured by each component.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define a transformation matrix
T = np.array([[3, 1],
              [0, 2]])

# Calculate eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(T)
e_val1, e_val2 = eigenvalues
e_vec1, e_vec2 = eigenvectors[:, 0], eigenvectors[:, 1]

print(f"Transformation Matrix T:\n{T}")
print(f"\nEigenvalue 1: {e_val1:.2f}, Eigenvector 1: {e_vec1}")
print(f"Eigenvalue 2: {e_val2:.2f}, Eigenvector 2: {e_vec2}")

# Create some vectors to transform
v1 = np.array([1, 1])     # A random vector
v2 = e_vec1               # The first eigenvector

# Apply the transformation T to the vectors
v1_transformed = T @ v1
v2_transformed = T @ v2

# --- Plotting ---
plt.figure(figsize=(10, 10))
ax = plt.gca()

# Plot original vectors
ax.arrow(0, 0, v1[0], v1[1], head_width=0.2, head_length=0.2, fc='blue', ec='blue', label='Original Vector v1')
ax.arrow(0, 0, v2[0], v2[1], head_width=0.2, head_length=0.2, fc='green', ec='green', label='Eigenvector v2')

# Plot transformed vectors
ax.arrow(0, 0, v1_transformed[0], v1_transformed[1], head_width=0.2, head_length=0.2, fc='cyan', ec='cyan', label='Transformed v1')
ax.arrow(0, 0, v2_transformed[0], v2_transformed[1], head_width=0.2, head_length=0.2, fc='lime', ec='lime', label='Transformed Eigenvector (v2 * eigval)')

ax.set_xlim(-1, 5)
ax.set_ylim(-1, 5)
ax.set_title("Visualizing Eigenvectors", fontsize=16)
ax.axhline(0, color='black',linewidth=0.5)
ax.axvline(0, color='black',linewidth=0.5)
ax.grid(True)
ax.legend()
plt.show()

print("\nNotice in the plot:")
print("The blue vector 'v1' changes direction when transformed.")
print("The green eigenvector 'v2' does NOT change direction, it is only scaled by its eigenvalue (2.0).")
```

3. Matrix Factorization: Uncovering Latent Features
This example uses Singular Value Decomposition (SVD), a powerful matrix factorization technique. In AI, SVD is used in recommendation systems to find latent features connecting users and items.

Concept: SVD breaks a large matrix (e.g., user-movie ratings) into three smaller matrices. These smaller matrices can represent "user-to-concept" relationships (U), the strength of those concepts (S), and "movie-to-concept" relationships (V). This is the basis for creating "embeddings."

```python
import numpy as np

# Create a sample user-item matrix (4 users, 5 movies)
# Values are ratings from 1-5. 0 means not rated.
ratings_matrix = np.array([
    [5, 5, 0, 1, 3],
    [4, 4, 1, 0, 2],
    [0, 1, 5, 4, 5],
    [1, 0, 4, 5, 4]
])

print("Original Ratings Matrix (Users x Movies):\n", ratings_matrix)

# Perform Singular Value Decomposition (SVD)
U, S_diag, Vh = np.linalg.svd(ratings_matrix)

# S is returned as a 1D array of singular values, we need to make it a diagonal matrix
S = np.zeros(ratings_matrix.shape)
S[:S_diag.shape[0], :S_diag.shape[0]] = np.diag(S_diag)

print("\n--- SVD Components ---")
print(f"\nU (User-to-Concept, {U.shape}):\n{np.round(U, 2)}")
print(f"\nS (Concept Strengths, {S.shape}):\n{np.round(S, 2)}")
print(f"\nVh (Movie-to-Concept, {Vh.shape}):\n{np.round(Vh, 2)}")

# To show it works, we can reconstruct the original matrix
reconstructed_matrix = U @ S @ Vh
print("\nReconstructed Matrix (should be close to original):\n", np.round(reconstructed_matrix, 1))

# By using only the top 'k' concepts, we can create a compressed approximation
k = 2
reconstructed_approx = U[:, :k] @ S[:k, :k] @ Vh[:k, :]
print(f"\nApproximation using top {k} concepts:\n", np.round(reconstructed_approx, 1))
print("\nThis approximation can be used to predict the missing (0) ratings.")
```

1. Differentiation: Partial Derivatives and Gradients
This example uses the sympy library for symbolic mathematics to compute partial derivatives and the gradient of a function. The gradient is a vector of partial derivatives, and it points in the direction of the steepest ascent of the function. In ML, we move in the opposite direction of the gradient to minimize loss.

Concept: For a loss function with many parameters (weights), we need to know how a small change in each parameter affects the total loss. This is what partial derivatives tell us. The gradient bundles this information together, forming the core of the backpropagation algorithm.

```python
import sympy as sp

# Define symbolic variables (parameters of a function)
x, y = sp.symbols('x y')

# Define a multivariable function (e.g., a simple loss function)
f = x**2 + 2*y**3

# --- Calculate Partial Derivatives ---
# Partial derivative with respect to x
df_dx = sp.diff(f, x)

# Partial derivative with respect to y
df_dy = sp.diff(f, y)

# --- Form the Gradient ---
# The gradient is a vector of the partial derivatives
gradient_vector = [df_dx, df_dy]

# --- Evaluate the gradient at a specific point, e.g., (x=2, y=1) ---
point = {x: 2, y: 1}
gradient_at_point = [deriv.evalf(subs=point) for deriv in gradient_vector]


print(f"Function f(x, y) = {f}")
print(f"Partial derivative with respect to x (âˆ‚f/âˆ‚x): {df_dx}")
print(f"Partial derivative with respect to y (âˆ‚f/âˆ‚y): {df_dy}")
print(f"Gradient of f (âˆ‡f): {gradient_vector}")
print(f"\nGradient evaluated at the point {point}: {gradient_at_point}")
```

2. Multivariable Calculus: Visualizing a Loss Function
This example plots a 3D "loss landscape" for a multivariable function. Understanding this surface is key to understanding optimization. We want to find the coordinates of the lowest point on this surface.

Concept: The parameters of a machine learning model (like the weights and biases of a neural network) define the dimensions of a high-dimensional landscape. The height of this landscape at any point is the model's error or "loss." Training the model is equivalent to finding the lowest valley in this landscape.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define a multivariable function representing a loss surface
def loss_function(w1, w2):
    return w1**2 + w2**2 + np.sin(w1) * 5

# Create a grid of w1, w2 values
w1_vals = np.linspace(-5, 5, 100)
w2_vals = np.linspace(-5, 5, 100)
w1_grid, w2_grid = np.meshgrid(w1_vals, w2_vals)

# Calculate the loss for each point on the grid
loss_grid = loss_function(w1_grid, w2_grid)

# --- Plotting the 3D surface ---
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(w1_grid, w2_grid, loss_grid, cmap='viridis', alpha=0.8)

ax.set_title("Multivariable Calculus: A 3D Loss Surface", fontsize=16)
ax.set_xlabel('Parameter 1 (w1)')
ax.set_ylabel('Parameter 2 (w2)')
ax.set_zlabel('Loss')
plt.show()

print("The plot shows a 'loss landscape'.")
print("The goal of optimization is to find the (w1, w2) coordinates of the lowest point.")
```

3. Integration: Calculating Probability
While less common than differentiation in daily ML practice, integration is fundamental to probability theory. This example shows how to calculate the probability of a random variable falling within a certain range by integrating its Probability Density Function (PDF).

Concept: For continuous variables, the probability of an event is not a single point but the area under the curve of the PDF over an interval. This is used in probabilistic models like Gaussian Mixture Models or when analyzing model uncertainty.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.integrate import quad

# Define a standard normal distribution (mean=0, std=1)
mu, sigma = 0, 1
x = np.linspace(-4, 4, 1000)
pdf = norm.pdf(x, mu, sigma)

# The range we want to find the probability for
a, b = -1.0, 1.0

# --- Integration to find the area under the curve ---
# quad returns the integral result and an estimate of the error
probability, error = quad(norm.pdf, a, b, args=(mu, sigma))

# --- Plotting ---
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, 'b-', label='Normal Distribution PDF')

# Shade the area that corresponds to the probability
x_fill = np.linspace(a, b, 100)
pdf_fill = norm.pdf(x_fill, mu, sigma)
plt.fill_between(x_fill, pdf_fill, color='cyan', alpha=0.6, label=f'Area from {a} to {b}')

plt.title("Integration for Probability Calculation", fontsize=16)
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.text(-3.8, 0.3, f'P({a} â‰¤ X â‰¤ {b}) = {probability:.4f}', fontsize=12)
plt.show()

print(f"The probability of a value falling between {a} and {b} is the integral of the PDF, which is {probability:.4f}")
print(f"(This is the familiar 68% rule for one standard deviation in a normal distribution)")
```

4. Optimization Techniques: Gradient Descent
This example implements the Gradient Descent algorithm from scratch. It navigates the 2D "loss landscape" from the second example by iteratively taking small steps in the opposite direction of the gradient.

Concept: Gradient Descent is the workhorse algorithm for training most ML models. It's how a model learns. By repeatedly calculating the gradient of the loss with respect to the model's parameters and updating them, the model systematically minimizes its error.

```python
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Define the loss function and its gradient ---
# Using the same function as the 3D plot
def loss_function(w):
    return w[0]**2 + w[1]**2 + np.sin(w[0]) * 5

def gradient(w):
    # Partial derivatives calculated manually or with sympy
    # âˆ‚f/âˆ‚w1 = 2*w1 + 5*cos(w1)
    # âˆ‚f/âˆ‚w2 = 2*w2
    return np.array([2*w[0] + 5*np.cos(w[0]), 2*w[1]])

# --- 2. Implement Gradient Descent ---
learning_rate = 0.1
n_iterations = 30
current_w = np.array([4.0, 4.0])  # A random starting point
path = [current_w]

for _ in range(n_iterations):
    grad = gradient(current_w)
    current_w = current_w - learning_rate * grad # The core update step
    path.append(current_w)

path = np.array(path)

# --- 3. Plotting the contour map and the descent path ---
w1_vals = np.linspace(-5, 5, 100)
w2_vals = np.linspace(-5, 5, 100)
w1_grid, w2_grid = np.meshgrid(w1_vals, w2_vals)
loss_grid = loss_function([w1_grid, w2_grid])

plt.figure(figsize=(10, 8))
plt.contourf(w1_grid, w2_grid, loss_grid, levels=20, cmap='viridis', alpha=0.7)
plt.colorbar(label='Loss')
plt.plot(path[:, 0], path[:, 1], 'ro-', label='Path of Gradient Descent')
plt.title("Optimization: Gradient Descent in Action", fontsize=16)
plt.xlabel('Parameter 1 (w1)')
plt.ylabel('Parameter 2 (w2)')
plt.legend()
plt.grid(True)
plt.show()

print(f"Starting point: {path[0].round(2)}")
print(f"Ending point (minimum found): {path[-1].round(2)}")
```

2. Bayes' Theorem
This example demonstrates Bayes' theorem with a classic medical diagnosis problem. It shows how to update our belief about a patient having a disease after receiving a positive test result.

Concept: Bayes' theorem is the heart of Bayesian inference and models like the Naive Bayes classifier. It provides a formal way to update a prior belief with new evidence (data) to arrive at a posterior belief.

$$ P(\text{Disease} \mid \text{Positive}) = \frac{P(\text{Positive} \mid \text{Disease}) \cdot P(\text{Disease})}{P(\text{Positive})} $$
â€‹

```python
# Scenario: A rare disease affects 1 in 10,000 people.
# A test for this disease is 99% accurate (true positive rate).
# The test also has a 0.5% false positive rate.
# Question: If a person tests positive, what is the actual probability they have the disease?

# --- 1. Define the probabilities ---
p_disease = 1 / 10000                            # Prior: P(Disease)
p_no_disease = 1 - p_disease                     # P(No Disease)

p_pos_given_disease = 0.99                       # Likelihood (True Positive): P(Positive | Disease)
p_pos_given_no_disease = 0.005                   # Likelihood (False Positive): P(Positive | No Disease)

# --- 2. Calculate P(Positive), the total probability of testing positive ---
# This is the 'evidence' term in the denominator of Bayes' theorem.
# P(Positive) = P(Positive | Disease) * P(Disease) + P(Positive | No Disease) * P(No Disease)
p_positive = (p_pos_given_disease * p_disease) + (p_pos_given_no_disease * p_no_disease)

# --- 3. Apply Bayes' Theorem ---
# P(Disease | Positive) = (P(Positive | Disease) * P(Disease)) / P(Positive)
p_disease_given_pos = (p_pos_given_disease * p_disease) / p_positive

print("--- Bayes' Theorem Example ---")
print(f"Prior probability of having the disease: {p_disease:.4%}")
print(f"Probability of a positive test (the evidence): {p_positive:.4%}")
print("\nAfter a positive test...")
print(f"The updated (posterior) probability of having the disease is: {p_disease_given_pos:.4%}")
print("\nThis surprising result is due to the rarity of the disease (low prior).")
```

4. Hypothesis Testing and Confidence Intervals
This example performs an independent t-test to determine if there is a statistically significant difference between the means of two groups. It also calculates the 95% confidence interval for the mean of one group.

Concept: Hypothesis testing is used to make decisions based on data. In ML, it can be used for A/B testing (e.g., does a new model perform better than an old one?) or feature selection. A confidence interval provides a range of plausible values for a population parameter (like the mean), quantifying the uncertainty in our estimate.

```python
from scipy import stats
import numpy as np

# Generate two groups of data (e.g., control vs. treatment)
np.random.seed(0)
group_A_scores = stats.norm.rvs(loc=5, scale=2, size=30) # Control group
group_B_scores = stats.norm.rvs(loc=5.8, scale=2, size=30) # Treatment group

# --- 1. Perform an Independent T-test ---
t_statistic, p_value = stats.ttest_ind(group_A_scores, group_B_scores)

# --- 2. Calculate 95% Confidence Interval for Group B ---
mean_b = np.mean(group_B_scores)
sem_b = stats.sem(group_B_scores) # Standard Error of the Mean
confidence_interval_b = stats.t.interval(0.95, len(group_B_scores)-1, loc=mean_b, scale=sem_b)

print("--- Hypothesis Test Results ---")
print(f"Mean of Group A: {np.mean(group_A_scores):.2f}")
print(f"Mean of Group B: {np.mean(group_B_scores):.2f}")
print(f"\nT-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# --- 3. Interpretation ---
alpha = 0.05
if p_value < alpha:
    print("\nConclusion: The difference is statistically significant (p < 0.05). We reject the null hypothesis.")
else:
    print("\nConclusion: The difference is not statistically significant (p >= 0.05). We fail to reject the null hypothesis.")

print("\n--- Confidence Interval ---")
print(f"The 95% confidence interval for the mean of Group B is: {confidence_interval_b[0]:.2f} to {confidence_interval_b[1]:.2f}")
print("This means we are 95% confident that the true population mean for Group B lies within this range.")
```

5. Maximum Likelihood Estimation (MLE)
This example demonstrates the core idea of MLE. We generate data from a Normal distribution with a known mean and standard deviation. Then, we write a function for the log-likelihood and use an optimizer to find the parameters (mean and std) that are most likely to have generated our observed data.

Concept: MLE is a foundational method for fitting a model's parameters to data. It answers the question: "What parameter values for my model make the data I observed the most probable?" Many ML models, like Logistic Regression, are trained using this principle.

```python
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt

# --- 1. Generate data from a known distribution ---
true_mu, true_sigma = 10, 2
np.random.seed(1)
data = np.random.normal(true_mu, true_sigma, 100)

# --- 2. Define the log-likelihood function for a Normal distribution ---
# We use the log-likelihood for numerical stability and easier optimization.
def log_likelihood_normal(params, data):
    mu, sigma = params[0], params[1]
    if sigma <= 0:
        return -np.inf # Return negative infinity for invalid sigma
    # Formula for log-likelihood of a normal distribution
    n = len(data)
    log_lik = -n/2 * np.log(2 * np.pi * sigma**2) - np.sum((data - mu)**2) / (2 * sigma**2)
    return log_lik

# --- 3. Maximize the log-likelihood using an optimizer ---
# We minimize the *negative* log-likelihood, which is equivalent to maximizing it.
neg_log_likelihood = lambda params, data: -log_likelihood_normal(params, data)

# Initial guess for the parameters
initial_guess = [5, 5]

# Use scipy's optimizer
result = minimize(neg_log_likelihood, initial_guess, args=(data,), method='L-BFGS-B', bounds=[(None, None), (1e-6, None)])

# Extract the estimated parameters
mle_mu, mle_sigma = result.x

print("--- Maximum Likelihood Estimation Results ---")
print(f"True parameters: mu = {true_mu}, sigma = {true_sigma}")
print(f"Sample mean (analytical MLE for mu): {np.mean(data):.4f}")
print(f"Sample std (analytical MLE for sigma): {np.std(data):.4f}")
print(f"\nParameters found by optimizer (MLE): mu = {mle_mu:.4f}, sigma = {mle_sigma:.4f}")

# --- Optional: Visualize the log-likelihood surface ---
mu_range = np.linspace(8, 12, 50)
sigma_range = np.linspace(1, 4, 50)
log_lik_surface = np.zeros((len(mu_range), len(sigma_range)))

for i, mu in enumerate(mu_range):
    for j, sigma in enumerate(sigma_range):
        log_lik_surface[j, i] = log_likelihood_normal([mu, sigma], data)

plt.figure(figsize=(9, 7))
plt.contourf(mu_range, sigma_range, log_lik_surface, levels=50, cmap='viridis')
plt.colorbar(label='Log-Likelihood')
plt.plot(true_mu, true_sigma, 'w*', markersize=15, label='True Parameters')
plt.plot(mle_mu, mle_sigma, 'r+', markersize=15, label='MLE Parameters')
plt.title('Log-Likelihood Surface', fontsize=16)
plt.xlabel('Mean ($\mu$)')
plt.ylabel('Standard Deviation ($\sigma$)')
plt.legend()
plt.show()
```

1. Convex vs. Non-Convex Optimization
This example plots two functions to illustrate the difference between a convex and a non-convex optimization problem. In machine learning, the "loss landscape" can be either, and its shape dramatically affects how easily a model can be trained.

Concept: A convex function has a single global minimum, like a perfect bowl. Any local minimum is also the global minimum, so algorithms like Gradient Descent are guaranteed to find the best solution. A non-convex function can have multiple local minima (small valleys) and a global minimum (the deepest valley). An algorithm might get stuck in a "good enough" local minimum and never find the best one. Deep learning loss landscapes are highly non-convex.

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Create subplots ---
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
fig.suptitle('Optimization Landscapes', fontsize=16)
x = np.linspace(-3, 3, 100)

# --- 1. Convex Function ---
# A simple quadratic function, like the one in Ordinary Least Squares regression.
def convex_function(x):
    return x**2

ax1.plot(x, convex_function(x))
ax1.set_title('Convex Function (One Global Minimum)')
ax1.set_xlabel('Parameter value (w)')
ax1.set_ylabel('Loss')
ax1.scatter(0, 0, color='red', s=100, zorder=5, label='Global Minimum')
ax1.legend()
ax1.grid(True)

# --- 2. Non-Convex Function ---
# A more complex polynomial, typical of deep learning loss surfaces.
def non_convex_function(x):
    return x**4 - 5*x**2 + x

# Find minima for plotting
local_min_x = 1.62
global_min_x = -1.67

ax2.plot(x, non_convex_function(x))
ax2.set_title('Non-Convex Function (Multiple Minima)')
ax2.set_xlabel('Parameter value (w)')
ax2.set_ylabel('Loss')
ax2.scatter(global_min_x, non_convex_function(global_min_x), color='red', s=100, zorder=5, label='Global Minimum')
ax2.scatter(local_min_x, non_convex_function(local_min_x), color='orange', s=100, zorder=5, label='Local Minimum')
ax2.legend()
ax2.grid(True)

plt.show()
```

2. Gradient-Based Methods: SGD vs. Adam
This example compares the optimization paths of standard Stochastic Gradient Descent (SGD) and the more advanced Adam optimizer on a challenging loss surface known as the Rosenbrock function.

Concept: SGD uses a fixed learning rate to follow the gradient. Adam (Adaptive Moment Estimation) adapts the learning rate for each parameter and uses "momentum" (an exponentially decaying average of past gradients) to accelerate descent and smooth out oscillations. In practice, optimizers like Adam are the default choice for training deep neural networks because they converge much faster and more reliably than simple SGD.

```python
import numpy as np
import matplotlib.pyplot as plt

# The Rosenbrock function is a classic non-convex optimization test case
def rosenbrock(w):
    x, y = w
    return (1 - x)**2 + 100 * (y - x**2)**2

def rosenbrock_grad(w):
    x, y = w
    grad_x = -2 * (1 - x) - 400 * x * (y - x**2)
    grad_y = 200 * (y - x**2)
    return np.array([grad_x, grad_y])

# --- Optimizer Implementations ---
def run_optimizer(optimizer_type='sgd', n_iterations=100):
    w = np.array([-1.5, -1.5]) # Starting point
    path = [w]

    # Optimizer-specific parameters
    lr = 0.001 if optimizer_type == 'sgd' else 0.02
    beta1, beta2, epsilon = 0.9, 0.999, 1e-8
    m, v = 0, 0

    for t in range(1, n_iterations + 1):
        grad = rosenbrock_grad(w)
        if optimizer_type == 'adam':
            m = beta1 * m + (1 - beta1) * grad
            v = beta2 * v + (1 - beta2) * (grad**2)
            m_hat = m / (1 - beta1**t)
            v_hat = v / (1 - beta2**t)
            update = lr * m_hat / (np.sqrt(v_hat) + epsilon)
        else: # SGD
            update = lr * grad

        w = w - update
        path.append(w)

    return np.array(path)

path_sgd = run_optimizer('sgd', n_iterations=300)
path_adam = run_optimizer('adam', n_iterations=300)

# --- Plotting ---
x = np.linspace(-2, 2, 200)
y = np.linspace(-2, 3, 200)
X, Y = np.meshgrid(x, y)
Z = rosenbrock([X, Y])

plt.figure(figsize=(12, 8))
plt.contour(X, Y, Z, levels=np.logspace(0, 5, 30), cmap='viridis')
plt.plot(path_sgd[:, 0], path_sgd[:, 1], 'o-', label='SGD Path', color='red', markersize=3)
plt.plot(path_adam[:, 0], path_adam[:, 1], 'o-', label='Adam Path', color='cyan', markersize=3)
plt.plot(1, 1, 'w*', markersize=15, label='Global Minimum') # Minimum is at (1,1)
plt.title('Optimizer Comparison: SGD vs. Adam', fontsize=16)
plt.xlabel('Parameter 1')
plt.ylabel('Parameter 2')
plt.legend()
plt.show()

print("Notice how SGD struggles in the long, narrow valley, while Adam's adaptive nature helps it navigate much more directly toward the minimum.")
```

3. Constrained and Unconstrained Optimization
This example uses scipy.optimize.minimize to find the minimum of a function first without constraints, and then with a linear constraint.

Concept: Unconstrained optimization seeks the minimum over the entire domain. Constrained optimization seeks the minimum only within a specific region or on a boundary. Regularization techniques in ML (like L1/L2) are a form of constrained optimization, where we minimize the loss subject to a constraint on the size of the model's weights to prevent overfitting.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Define a simple objective function to minimize
def objective_function(w):
    return (w[0] - 2)**2 + (w[1] - 3)**2

# --- 1. Unconstrained Optimization ---
# Find the minimum of the function anywhere
initial_guess = [0, 0]
unconstrained_result = minimize(objective_function, initial_guess)
w_unconstrained = unconstrained_result.x

# --- 2. Constrained Optimization ---
# Now, find the minimum subject to the constraint w[0] + w[1] = 1
constraint = {'type': 'eq', 'fun': lambda w: w[0] + w[1] - 1}
constrained_result = minimize(objective_function, initial_guess, constraints=[constraint])
w_constrained = constrained_result.x

print("--- Optimization Results ---")
print(f"Unconstrained Minimum found at: {w_unconstrained.round(2)}")
print(f"Constrained Minimum found at:   {w_constrained.round(2)}")

# --- Plotting ---
w1 = np.linspace(-1, 5, 200)
w2 = np.linspace(-1, 5, 200)
W1, W2 = np.meshgrid(w1, w2)
Z = objective_function([W1, W2])

plt.figure(figsize=(9, 8))
plt.contour(W1, W2, Z, levels=15, cmap='viridis')
plt.plot(w_unconstrained[0], w_unconstrained[1], 'ro', markersize=10, label='Unconstrained Minimum')
plt.plot(w_constrained[0], w_constrained[1], 'yo', markersize=10, label='Constrained Minimum')

# Plot the constraint line w1 + w2 = 1
plt.plot(w1, 1 - w1, 'b--', label='Constraint: w1 + w2 = 1')

plt.title('Constrained vs. Unconstrained Optimization', fontsize=16)
plt.xlabel('Parameter 1 (w1)')
plt.ylabel('Parameter 2 (w2)')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```

4. Lagrange Multipliers (Conceptual Visualization)
This example visually demonstrates the core principle of Lagrange multipliers: at the optimal point of a constrained problem, the gradient of the objective function is parallel to the gradient of the constraint function.

Concept: The Lagrange multiplier (
lambda) is a scalar that scales the gradient of the constraint function so that it exactly matches the gradient of the objective function at the solution. This method provides the formal mathematical basis for solving the kind of constrained problems shown in the previous example, which are common in models like Support Vector Machines (SVMs).

```python
import numpy as np
import matplotlib.pyplot as plt

# Using the functions and results from the previous example
def objective_grad(w):
    return np.array([2 * (w[0] - 2), 2 * (w[1] - 3)])

def constraint_grad(w):
    # Constraint g(w) = w[0] + w[1] - 1 = 0
    # âˆ‡g = [âˆ‚g/âˆ‚w1, âˆ‚g/âˆ‚w2]
    return np.array([1, 1])

# The constrained solution found previously
w_opt = np.array([-1.0, 2.0]) # From the previous example, the solution is ( -1, 2)
# No, that's not right. The solution from scipy was w = [0., 1.]
w_opt = w_constrained # This is the correct solution from the scipy example above. w_opt is [0, 1]

# Calculate the gradients at the optimal point
grad_f_at_opt = objective_grad(w_opt)
grad_g_at_opt = constraint_grad(w_opt)

# In the theory, âˆ‡f = Î»âˆ‡g. So Î» = ||âˆ‡f|| / ||âˆ‡g|| (or component-wise)
# Let's find lambda from the first component: âˆ‡f_x = Î» * âˆ‡g_x => -4 = Î» * 1 => Î» = -4
lambda_val = grad_f_at_opt[0] / grad_g_at_opt[0]

print("--- Lagrange Multiplier Principle ---")
print(f"Optimal point w*: {w_opt.round(2)}")
print(f"Gradient of objective at w* (âˆ‡f): {grad_f_at_opt.round(2)}")
print(f"Gradient of constraint at w* (âˆ‡g): {grad_g_at_opt.round(2)}")
print(f"Calculated Lagrange multiplier (Î»): {lambda_val:.2f}")
print(f"Check: Î» * âˆ‡g = {lambda_val * grad_g_at_opt}, which should equal âˆ‡f.")

# --- Plotting ---
w1 = np.linspace(-1, 5, 200)
w2 = np.linspace(-1, 5, 200)
W1, W2 = np.meshgrid(w1, w2)
Z = objective_function([W1, W2])

plt.figure(figsize=(9, 8))
plt.contour(W1, W2, Z, levels=15, cmap='viridis', alpha=0.7)
plt.plot(w_opt[0], w_opt[1], 'yo', markersize=10, label='Optimal Point')
plt.plot(w1, 1 - w1, 'b--', label='Constraint Line')

# Plot the gradients as arrows
plt.arrow(w_opt[0], w_opt[1], grad_f_at_opt[0], grad_f_at_opt[1],
          color='red', head_width=0.2, length_includes_head=True, label='âˆ‡f (Objective Gradient)')
plt.arrow(w_opt[0], w_opt[1], grad_g_at_opt[0], grad_g_at_opt[1],
          color='purple', head_width=0.2, length_includes_head=True, label='âˆ‡g (Constraint Gradient)')

plt.title('Lagrange Multiplier Principle Visualization', fontsize=16)
plt.xlabel('Parameter 1 (w1)')
plt.ylabel('Parameter 2 (w2)')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()

print("\nNotice how the red and purple arrows are parallel. This shows that the gradients are aligned, which is the condition for optimality in a constrained problem.")
```

2. Graph Theory
This example uses the networkx library to create a simple graph, representing, for example, a road network or connections in a social network. It then finds and visualizes the shortest path between two nodes.

Concept: Graphs are used to model relationships between entities. Graph Neural Networks (GNNs) operate directly on graph structures to make predictions (e.g., classifying users in a social network). Knowledge graphs store information as nodes and relationships, which is central to modern search and question-answering systems. Pathfinding algorithms are core to reinforcement learning and logistics problems.

```python
import networkx as nx
import matplotlib.pyplot as plt

# --- 1. Create a weighted graph ---
# This could represent cities (nodes) and driving times (weights)
G = nx.Graph()
edges = [
    ('A', 'B', {'weight': 1}), ('A', 'C', {'weight': 4}), ('B', 'C', {'weight': 2}),
    ('B', 'D', {'weight': 5}), ('C', 'D', {'weight': 1}), ('C', 'E', {'weight': 3}),
    ('D', 'E', {'weight': 1}), ('D', 'F', {'weight': 2}), ('E', 'F', {'weight': 2})
]
G.add_edges_from(edges)

# --- 2. Find the shortest path using Dijkstra's algorithm ---
start_node = 'A'
end_node = 'F'
shortest_path = nx.dijkstra_path(G, source=start_node, target=end_node, weight='weight')
shortest_path_length = nx.dijkstra_path_length(G, source=start_node, target=end_node, weight='weight')

print(f"Shortest path from {start_node} to {end_node}: {shortest_path}")
print(f"Total weight (distance) of shortest path: {shortest_path_length}")

# --- 3. Plotting the graph and the path ---
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G, seed=42) # Positions for all nodes

# Draw all nodes and edges
nx.draw_networkx_nodes(G, pos, node_size=700, node_color='skyblue')
nx.draw_networkx_labels(G, pos, font_size=12, font_color='black')
nx.draw_networkx_edges(G, pos, width=1.0, alpha=0.5)

# Highlight the shortest path
path_edges = list(zip(shortest_path, shortest_path[1:]))
nx.draw_networkx_edges(G, pos, edgelist=path_edges, width=3, alpha=0.8, edge_color='r')

# Draw edge labels (weights)
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

plt.title('Graph Theory: Shortest Path Analysis', fontsize=16)
plt.axis('off')
plt.show()
```

##**Python examples that illustrate the key topics of Information Theory and their application in advanced areas of AI and Machine Learning.**

1. Entropy, Mutual Information, and KL Divergence
This example covers three fundamental concepts from information theory.

Entropy: Measures the average level of "surprise" or uncertainty inherent in a variable's possible outcomes. A higher entropy means more uncertainty.

KL Divergence: Measures how one probability distribution P diverges from a second, expected probability distribution Q. It's a measure of the information lost when Q is used to approximate P.

Mutual Information: Measures the mutual dependence between two variables. It quantifies the information obtained about one random variable by observing the other. It is used in feature selection to find features that are most informative about the target class.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import entropy
from sklearn.metrics import mutual_info_score

# --- 1. Entropy: Quantifying Uncertainty ---
print("--- 1. Entropy ---")
# A fair coin has high uncertainty (maximum entropy)
prob_fair_coin = np.array([0.5, 0.5])
entropy_fair = entropy(prob_fair_coin, base=2)
print(f"Probability distribution (Fair Coin): {prob_fair_coin}")
print(f"Entropy of a fair coin: {entropy_fair:.4f} bits")

# A biased coin has low uncertainty (lower entropy)
prob_biased_coin = np.array([0.9, 0.1])
entropy_biased = entropy(prob_biased_coin, base=2)
print(f"\nProbability distribution (Biased Coin): {prob_biased_coin}")
print(f"Entropy of a biased coin: {entropy_biased:.4f} bits")


# --- 2. Kullback-Leibler (KL) Divergence: Measuring Distance Between Distributions ---
print("\n--- 2. KL Divergence ---")
# This is also known as relative entropy.
# It measures the information lost when approximating P with Q.
x = np.linspace(-10, 10, 1000)
# Distribution P (the "true" distribution)
p_dist = 1/np.sqrt(2*np.pi*4) * np.exp(-(x-0)**2/(2*4))
# Distribution Q (our model's approximation)
q_dist = 1/np.sqrt(2*np.pi*9) * np.exp(-(x-2)**2/(2*9))

# Normalize to ensure they sum to 1
p_dist_norm = p_dist / np.sum(p_dist)
q_dist_norm = q_dist / np.sum(q_dist)

# Calculate KL Divergence D_KL(P || Q)
kl_pq = entropy(p_dist_norm, qk=q_dist_norm)
print(f"KL Divergence D_KL(P || Q): {kl_pq:.4f}")
print("This value indicates how much information is lost when using Q to represent P.")

# Plotting the distributions
plt.figure(figsize=(10, 6))
plt.plot(x, p_dist_norm, 'b-', label='Distribution P (True)')
plt.plot(x, q_dist_norm, 'r--', label='Distribution Q (Model)')
plt.title('KL Divergence Visualization', fontsize=16)
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()


# --- 3. Mutual Information: Measuring Dependency ---
print("\n--- 3. Mutual Information ---")
# Use case: Feature selection. Does knowing a feature's value reduce uncertainty about the target class?
# Create two categorical variables. Let's say 'weather' and 'play_tennis'.
# In this data, when it's sunny, you are more likely to play tennis.
weather =      ['Sunny', 'Sunny', 'Overcast', 'Rainy', 'Rainy', 'Rainy', 'Overcast', 'Sunny']
play_tennis =  ['Yes',   'No',    'Yes',      'Yes',   'No',    'No',    'Yes',      'Yes']

mi_score = mutual_info_score(weather, play_tennis)
print(f"Mutual Information between 'weather' and 'play_tennis': {mi_score:.4f}")
print("A score > 0 means the variables are not independent; knowing one gives you information about the other.")

# Example with an irrelevant feature
age = [20, 25, 30, 35, 40, 45, 50, 55]
mi_score_irrelevant = mutual_info_score(age, play_tennis)
print(f"Mutual Information between 'age' and 'play_tennis': {mi_score_irrelevant:.4f}")
print("A score of 0.0 means the variables are independent.")
```

2. Cross-Entropy Loss
This example demonstrates how the Cross-Entropy Loss function works for a multi-class classification problem. It is the most common loss function for classification tasks in deep learning.

Concept: Cross-entropy measures the "distance" between the true probability distribution (a one-hot vector where the correct class is 1 and all others are 0) and the predicted probability distribution from the model (the output of a softmax layer). The goal of training is to minimize this distance, pushing the model's predictions to be as close to the true labels as possible.

The formula for a single sample is: L=âˆ’
sum_iy_i
log(
haty_i), where y_i is the true probability for class i, and
haty_i is the predicted probability.

```python
import numpy as np

def cross_entropy_loss(y_true, y_pred):
    """
    Calculates the cross-entropy loss for a single prediction.
    :param y_true: A list or numpy array of true labels (one-hot encoded).
    :param y_pred: A list or numpy array of predicted probabilities.
    :return: The cross-entropy loss (a scalar).
    """
    # Adding a small epsilon to avoid log(0) which is undefined
    epsilon = 1e-15
    y_pred = np.clip(y_pred, epsilon, 1. - epsilon)
    return -np.sum(y_true * np.log(y_pred))

# --- Scenario: Multi-class classification with 3 classes (e.g., Cat, Dog, Bird) ---

# The true label is 'Dog'. In one-hot encoding, this is [0, 1, 0].
y_true = np.array([0, 1, 0])

# --- Case 1: Good Prediction ---
# The model is confident and correct. It assigns a high probability to the 'Dog' class.
y_pred_good = np.array([0.1, 0.8, 0.1]) # Probabilities from the model's softmax layer
loss_good = cross_entropy_loss(y_true, y_pred_good)

print("--- Cross-Entropy Loss Example ---")
print(f"\nTrue Label (one-hot): {y_true}")
print("\n--- Case 1: Good Prediction ---")
print(f"Model Predictions: {y_pred_good}")
print(f"Cross-Entropy Loss: {loss_good:.4f}")


# --- Case 2: Bad Prediction ---
# The model is confident but WRONG. It assigns a high probability to the 'Cat' class.
y_pred_bad = np.array([0.7, 0.2, 0.1])
loss_bad = cross_entropy_loss(y_true, y_pred_bad)

print("\n--- Case 2: Bad Prediction ---")
print(f"Model Predictions: {y_pred_bad}")
print(f"Cross-Entropy Loss: {loss_bad:.4f}")

# --- Case 3: Uncertain Prediction ---
# The model is not sure and assigns roughly equal probability to all classes.
y_pred_uncertain = np.array([0.3, 0.4, 0.3])
loss_uncertain = cross_entropy_loss(y_true, y_pred_uncertain)

print("\n--- Case 3: Uncertain Prediction ---")
print(f"Model Predictions: {y_pred_uncertain}")
print(f"Cross-Entropy Loss: {loss_uncertain:.4f}")

print("\nNotice how the loss is lowest for the good prediction and highest for the confidently wrong prediction.")
print("This is exactly the behavior we want from a loss function to guide model training.")
```
